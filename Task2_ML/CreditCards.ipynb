```python
X = df.iloc[:, :-1]
y = df.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=84, stratify=y)
```

будем идти по мере усложнения моделей (причём как без сэмплирования, так и с ним + везде стандартизируем данные), а по метрикам будем единовременно смотреть precision и recall и f2 метрику (почему не f1 - потому что важнее всё же отделять мошеннические операции)

соответственно порядок действий следующий

1) KNN
2) KNN с сэмплированием
3) Логистическая регрессия (исключительно без сэмплирования, потому что на ней сэмплирование даёт некорректный результат)
4) Рандомный лес (простые деревья переобучатся, тк много фичей, а лес в целом должен с задачей справиться)
5) Рандомный лес с сэмплированием

критерии итогового выбора модели - максимальные precision и recall (приоритет отдаётся precision, тк опять же - важнее отделять мошеннические операции)

SVC чисто из принципа использован не будет, потому что на больших датасетах (как наш) он очень долго будет работать

Тут обязательно нужно делать пересэмплирование, тк мош. операций мало

```python
# для f2 score

betta = 2

f2_lambda = lambda pr, re, betta: (1 + np.power(betta, 2)) * (pr * re) / (np.power(betta, 2) * pr + re)
```

```python
# соседей ставим поменьше, тк один из классов небольшой сам по себе

from sklearn.pipeline import Pipeline

pipe_knn = Pipeline([('scaler', StandardScaler()), 
                     ('clf', KNeighborsClassifier(n_neighbors=5))])

pipe_knn.fit(X_train, y_train)
y_pred_knn = pipe_knn.predict(X_test)

precision = precision_score(y_test, y_pred_knn)
recall = recall_score(y_test, y_pred_knn)

print(f'precision: {precision}, recall: {recall}, f2-score: {f2_lambda(precision, recall, betta)}')
```

```python
# честно сказать, не нашёл, как в sklearn-овский Pipeline вносить сэмплирование, 
# поэтому каждый раз буду 💫импортировать заново💫  

from imblearn.pipeline import Pipeline 

pipe_knn_sample = Pipeline([('scaler', StandardScaler()), 
                            ('adasyn', ADASYN(random_state=84)),
                            ('clf', KNeighborsClassifier(n_neighbors=5))])

pipe_knn_sample.fit(X_train, y_train)
y_pred_knn_sample = pipe_knn_sample.predict(X_test)

precision = precision_score(y_test, y_pred_knn_sample)
recall = recall_score(y_test, y_pred_knn_sample)

print(f'precision: {precision}, recall: {recall}, f2-score: {f2_lambda(precision, recall, betta)}')
```

```python
# class_weight = 'balanced' ставится для использования модели с несбалансированными данными

from sklearn.pipeline import Pipeline

pipe_ll = Pipeline([('scaler', StandardScaler()), 
                    ('clf', LogisticRegression(class_weight = 'balanced', n_jobs=-1))])

pipe_ll.fit(X_train, y_train)
y_pred_ll = pipe_ll.predict(X_test)

precision = precision_score(y_test, y_pred_ll)
recall = recall_score(y_test, y_pred_ll)

print(f'precision: {precision}, recall: {recall}, f2-score: {f2_lambda(precision, recall, betta)}')
```

```python
pipe_rf = Pipeline([('scaler', StandardScaler()), 
                    ('clf', RandomForestClassifier(class_weight = 'balanced', n_jobs=-1))])

pipe_rf.fit(X_train, y_train)
y_pred_rf = pipe_rf.predict(X_test)

precision = precision_score(y_test, y_pred_rf)
recall = recall_score(y_test, y_pred_rf)

print(f'precision: {precision}, recall: {recall}, f2-score: {f2_lambda(precision, recall, betta)}')
```

```python
from imblearn.pipeline import Pipeline 

pipe_rf_sample = Pipeline([('scaler', StandardScaler()),
                           ('adasyn', ADASYN(random_state=84)),
                           ('clf', RandomForestClassifier(n_jobs=-1))])

pipe_rf_sample.fit(X_train, y_train)
y_pred_rf_sample = pipe_rf_sample.predict(X_test)

precision = precision_score(y_test, y_pred_rf_sample)
recall = recall_score(y_test, y_pred_rf_sample)

print(f'precision: {precision}, recall: {recall}, f2-score: {f2_lambda(precision, recall, betta)}')
```

по итогу по метрикам два лидера - KNeighborsClassifier и RandomForestClassifier без сэмплирования 

думаю, здесь лучше оставиться на KNeighborsClassifier, поскольку у него больший precision (а мош. операции отделять важнее)
